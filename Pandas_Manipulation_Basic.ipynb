# 3. FILTERING

## Value Counts

````python
#Notice: using a SERIES here
df['Employment.Type'].value_counts()
````

## Filter 1 dimension

< less than


.> more than

<= less than or equal


.>= more than or equal

== equal

!= not equal

````python
#Define a filter
#It is a series of boolan values T/F 
employment_filter = df['Employment.Type'] == 'Salaried'
````

````python
employment_filter.head()
````

````python
#Filter observations with the defined filter
#Choose all colums with observations satisfying the filter

df.loc[employment_filter, :].sample(5)
````

````python
#Assign a dataframe with the filtered result

df_new = df.loc[employment_filter, :]
````

## Filter 2 or more dimensions

& = and 

| = or

^ = exclusive or

~  = not


````python
# 2nd filter

ltv_filter = df['ltv'] < 100
````

````python
# Combine 2 filters with &

df_new = df.loc[employment_filter & ltv_filter, :]
````

````python
df_new.sample(5)
````

#6. MISSING DATA

## Identify missing data

It can be in a form of
1. Zero
2. False
3. NA - not applicable 
4. Empty strings

````python
#Identify no of missing data
df.info()

#Line 9: employment_type has missing values
````

````python
#Create a variable to isolate missing data

missing_employment_type = df['Employment_type'].isna()
````

````python
#Rows with missing value of Employment types

df.loc[missing_employment_type, :].head()
````

````python
# Rows without missing values, using ~
# Can assign a dataframe as a listwise method for missing data
df.loc[~missing_employment_type, :].head()
````

````python
# Counting NA or observations with NA

df['Employment_type'].isna().sum()
````

## Hanlding missing values
1. Using `.dropna` method (listwise)
2. Imputating values (fill) with `.fillna()`,
    `bfill()` or `ffill()` in Timeseries, `.interpolate(method ='linear')`,

````python
#Withtin observations 30 -39, delete any observation with NA value
df[30:40].dropna(how = 'any')
````

````python
# Using before fill .bfill() method for imputation 

df['Employment_type'].bfill()[8:10]
````

````python
#Fill values in observations with missing values

df['Employment_type'] = df['Employment_type'].bfill()
````

````python
df['Employment_type'].isna().sum()
````
